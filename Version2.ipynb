### MAE VS RMSE

* MAE : used for small errors
- RMSE : for big valued errors

|Criterion	|Mean Absolute Error (MAE)	|Root Mean Squared Error (RMSE)|
|-------------|---------------------------------|-----------------------|
|Formula	|$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n}$|	$y_i - \hat{y}_i$|
|Sensitivity to Outliers	|Less sensitive to outliers because it uses the absolute value of the errors.|	More sensitive to outliers because it squares the errors before averaging.|
|Interpretation	|Easier to interpret as it represents the average magnitude of the errors in the same units as the original data.	|More difficult to interpret because it is in the units of the original data but is influenced by the square of the errors.|
Use Cases	|Preferred when the cost of errors is proportional to the size of the error.	|Preferred when large errors are particularly undesirable.|
|Mathematical Properties	|Does not assume any specific distribution of the errors.	|Assumes that the errors are normally distributed.|
|Computational Complexity	|Generally simpler and faster to compute.	|More computationally intensive due to the square root and squaring operations.|

### Decision Tree vs Random forest regression


|Criterion	|Decision Tree Regression	|Random Forest Regression|
|---------------|--------------------------|------------------------|
|Model Structure	|Single decision tree.	|Ensemble of multiple decision trees.|
|Training Process	|Splits the data based on feature values to minimize a cost function (e.g., mean squared error).	|Combines the predictions of multiple decision trees, each trained on a different subset of the data.|
|Overfitting	|High risk of overfitting, especially with deep trees.	|Reduces overfitting by averaging the predictions of multiple trees.|
|Bias-Variance Tradeoff	|High variance, low bias.	|Lower variance, slightly higher bias compared to a single decision tree.|
|Feature Importance	|Can provide feature importance based on the splits.	|rovides more robust feature importance by averaging across multiple trees.|
|Prediction Stability	|Predictions can be unstable and sensitive to small changes in the data.	|More stable predictions due to the averaging effect of multiple trees.
|Computational Complexity	|Generally faster to train and predict.	|More computationally intensive due to the training of multiple trees.|
|Hyperparameters	|Max depth, min samples split, min samples leaf.	|Number of trees, max depth, min samples split, min samples leaf, max features.|
|Interpretability	|Easier to interpret as a single tree can be visualized.	|Less interpretable due to the ensemble nature, but feature importance can be analyzed.|
|Use Cases	|Suitable for small datasets or when interpretability is crucial.	|Suitable for larger datasets and when robustness and stability are more important.|

### Linear regression

Linear regression is a statistical model that estimates the linear relationship between a scalar response (dependent variable) and one or more explanatory variables (independent variables). The model aims to find the best-fitting line that minimizes the sum of the squared errors between observed and predicted values.

$Y = mX + b$,

where $Y$ is the dependent variable, $X$ is the independent variable, $m$ is the slope, and $b$ is the intercept.



### Polynomial regression

Polynomial regression is a form of linear regression in which the relationship between the independent variable x and dependent variable y is modeled as an nth-degree polynomial. It is used to model non-linear relationships between the variables.

In polynomial regression, the relationship between the variables is modeled using a polynomial equation of degree n. The equation takes the form:

$y = a + bx + cx^2 + dx^3 + … + nx^n$

where y is the dependent variable, x is the independent variable, and a, b, c, d, …, n are the coefficients of the polynomial.



```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate some sample data
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([50, 55, 65, 80, 110, 150, 200])

# Create a polynomial feature transformer
poly_transformer = PolynomialFeatures(degree=2)

# Transform the data
x_poly = poly_transformer.fit_transform(x.reshape(-1, 1))

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(x_poly, y)

# Print the coefficients of the polynomial
print(model.coef_)
```

### Residual sum of squares

The Residual Sum of Squares (RSS) is a statistical measure that evaluates the discrepancy between observed data values and their predicted values based on a regression model. It represents the sum of the squared differences between the actual data points and their corresponding predicted values.

Formula:

$RSS = Σ(ei)^2$

where:
$RSS$ is the Residual Sum of Squares
$ei$ is the residual (or error) for each data point, calculated as the difference between the observed value (yi) and the predicted value (ŷi) from the regression model: ei = yi - ŷi
$Σ$ denotes the summation of the squared residuals over all data points

**Properties**:
Non-negativity: RSS is always non-negative, as the sum of squared residuals cannot be negative.
Additivity: RSS is additive, meaning that the sum of the squared residuals for multiple regression models can be calculated by summing the RSS values for each model.
Homoscedasticity: RSS assumes that the variance of the residuals is constant across all data points, a condition known as homoscedasticity.

```python
def count_outliers(columns_to_count):
    outlier_counts = {}

    for col in columns_to_count:
        Q1 = housing[col].quantile(0.25)
        Q3 = housing[col].quantile(0.75)
    
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
    
        outliers = housing[(housing[col] < lower_bound) | (housing[col] > upper_bound)]
        outlier_counts[col] = len(outliers)

    print("Display the number of outliers for each column")
    print(outlier_counts)

    return outlier_counts
```

## Machine learning

```python
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np

housing["income_cat"] = np.ceil(housing["median_income"]/1.5) 
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True) # replace where false 

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) 

for train_index, test_index in split.split(housing, housing["income_cat"]): 
 strat_train_set = housing.loc[train_index] 
 strat_test_set = housing.loc[test_index] 

for set_ in (strat_train_set, strat_test_set): 
 set_.drop("income_cat", axis=1, inplace=True) 

housing = strat_train_set.copy()
```

```python
housing = strat_train_set.drop("median_house_value", axis=1) 
housing_labels = strat_train_set["median_house_value"].copy() 

from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler 
from sklearn.base import BaseEstimator, TransformerMixin 
from sklearn.compose import ColumnTransformer 

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6 

class CombinedAttributesAdder(BaseEstimator, TransformerMixin): 
 def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs 
    self.add_bedrooms_per_room = add_bedrooms_per_room 
 def fit(self, X, y=None): 
    return self # nothing else to do 
 def transform(self, X, y=None): 
    rooms_per_household = X[:, rooms_ix] / X[:, household_ix] 
    population_per_household = X[:, population_ix] / X[:, household_ix] 
    if self.add_bedrooms_per_room: 
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix] 
        return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room] 
    else: 
        return np.c_[X, rooms_per_household, population_per_household] 

num_pipeline = Pipeline([ 
 ('imputer', SimpleImputer(strategy="median")), 
 ('attribs_adder', CombinedAttributesAdder()), 
 ('std_scaler', StandardScaler()), 
 ]) 

housing_num = housing.select_dtypes(include=[np.number]) 
num_attribs = list(housing_num) 
cat_attribs = ["ocean_proximity"] 

full_pipeline = ColumnTransformer([ 
 ("num", num_pipeline, num_attribs), 
 ("cat", OneHotEncoder(), cat_attribs), 
 ]) 

housing_prepared = full_pipeline.fit_transform(housing)
```

```python
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error 
from sklearn.ensemble import RandomForestRegressor 

housing_true = housing_labels
forest_reg = RandomForestRegressor(random_state=42, n_estimators=800)
forest_reg.fit(housing_prepared, housing_labels)

housing_predictions = forest_reg.predict(housing_prepared)
forest_rmse = mean_squared_error(housing_true, housing_predictions)
forest_rmse = np.sqrt(forest_rmse)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score

def display_scores(scores): 
    print("Scores:", scores) 
    print("Mean:", scores.mean()) 
    print("Standard deviation:", scores.std()) 
```

```python
forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels, 
 scoring="neg_mean_squared_error", cv=10) 
forest_rmse_scores = np.sqrt(-forest_scores) 
display_scores(forest_rmse_scores)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint 

param_distribs = { 
 'n_estimators': randint(low=1, high=200), 
 'max_features': randint(low=1, high=8), 
 }
random_search = RandomizedSearchCV(param_distribs, n_iter=10, cv=5, scoring='neg_mean_squared_error',return_train_score=True, random_state=42) 
random_search.fit(housing_prepared, housing_labels)
```
