### Perceptron vs Logistic regression 
The Perceptron is a type of artificial neuron that uses a step function as the activation function. It is a basic procedure based on dot products between examples and weights, while Logistic regression is a linear model that uses a logistic function to predict probabilities. It is a type of supervised machine-learning algorithm used for classification tasks. The output from logistic regression is a probability between 0 and 1, which can be used to predict the class label. 

|Criterion	|Perceptron	|Logistic Regression|
|-----------|-----------|-------------------|
|Model Type	|Linear classifier.	|Probabilistic linear classifier.|
|Output	Binary output |(0 or 1).	|Probability output (between 0 and 1).|
|Decision Boundary	|Linear decision boundary.	|Linear decision boundary.|
|Loss Function	|Perceptron loss (hinge loss for binary classification).	|Logistic loss (cross-entropy loss).|
|Optimization	|Gradient descent or stochastic gradient descent.	|Gradient descent, stochastic gradient descent, or other optimization algorithms.|
|Convergence	|Converges if the data is linearly separable.	|Converges regardless of linear separability.|
Probabilistic Interpretation	|No probabilistic interpretation.	|Provides a probabilistic interpretation of the output.|
|Handling of Non-linearity	|Cannot handle non-linear decision boundaries.	|Cannot handle non-linear decision boundaries directly, but can be extended with kernel tricks or feature engineering.|
|Regularization	|Typically no regularization.	|Can include regularization terms (L1, L2) to prevent overfitting.|
|Use Cases	|Simple binary classification problems.	|Binary classification problems where probabilistic outputs are useful, such as in medical diagnosis or spam detection.|
|Interpretability	|Simple and interpretable as a linear model.	|Simple and interpretable as a linear model, with the added benefit of probabilistic outputs.
|Computational Complexity	|Generally faster and simpler to train.	|Slightly more computationally intensive due to the logistic function.|

### Maximum margin vs Linear classifiers


|Criterion	|Maximum Margin Classifiers (e.g., SVM)	|Linear Classifiers (e.g., Perceptron, Logistic Regression)|
|-----------|---------------------------------------|----------------------------------------------------------|
|Objective	|Maximize the margin between the decision boundary and the nearest data points (support vectors).	|Minimize a loss function (e.g., hinge loss for Perceptron, cross-entropy loss for Logistic Regression).|
Decision Boundary	|Linear or non-linear decision boundary (using kernel tricks).	|Linear decision boundary.
Margin	|Explicitly maximizes the margin.	|Does not explicitly maximize the margin.|
Loss Function	|Hinge loss or other margin-based losses.	|Various loss functions (e.g., hinge loss for Perceptron, cross-entropy loss for Logistic Regression).|
Optimization	|Quadratic programming or other optimization techniques.	|Gradient descent, stochastic gradient descent, or other optimization algorithms.|
Handling of Non-linearity	|Can handle non-linear decision boundaries using kernel tricks.	|Cannot handle non-linear decision boundaries directly, but can be extended with feature engineering.|
Regularization	|Inherent regularization through margin maximization.	|Can include regularization terms (L1, L2) to prevent overfitting.|
Support Vectors	|Only a subset of the data points (support vectors) influence the decision boundary.	|All data points influence the decision boundary.|
Robustness to Outliers	|More robust to outliers due to margin maximization.	|Less robust to outliers, especially in the case of Perceptron.|
Probabilistic Interpretation	|No probabilistic interpretation.	|Logistic Regression provides a probabilistic interpretation of the output.|
|Use Cases	|Suitable for high-dimensional spaces and cases where the margin is important.	|Suitable for binary classification problems where probabilistic outputs are useful or simplicity is preferred.
|Interpretability	|Less interpretable due to the focus on support vectors and potential use of kernels.	|More interpretable as linear models, especially Logistic Regression with probabilistic outputs.
|Computational Complexity	|More computationally intensive, especially with non-linear kernels.	|Generally faster and simpler to train, especially for linear models.|

### KNN 

The K-Nearest Neighbors (KNN) algorithm is a non-parametric, supervised machine learning method used for both classification and regression tasks. It was first developed by Evelyn Fix and Joseph Hodges in 1951 and later expanded by Thomas Cover.

#### Key Concepts
K: The number of nearest neighbors to consider when making a prediction.
Distance metric: The method used to calculate the distance between data points, such as Euclidean distance, Manhattan distance, or Minkowski distance.
Similarity: The algorithm assumes that similar items are close to each other in a feature space.

Example of k-NN classification. The test sample (green dot) should be classified either to blue squares or to red triangles. If k = 3 (solid line circle) it is assigned to the red triangles because there are 2 triangles and only 1 square inside the inner circle. If k = 5 (dashed line circle) it is assigned to the blue squares (3 squares vs. 2 triangles inside the outer circle).


#### How KNN Works

* Training phase: The algorithm stores all available training data points and their corresponding labels (class or value).
- Query phase: When a new, unseen data point (query) is input, the algorithm finds the K nearest neighbors to the query point based on the chosen distance metric.
* Prediction: The algorithm predicts the class or value of the query point by aggregating the labels of its K nearest neighbors. Common aggregation methods include:
  - Majority vote: Assign the class or value of the majority of the K nearest neighbors.
  * Weighted average: Calculate a weighted average of the values of the K nearest neighbors, where the weights are typically based on the distance from the query point.

|Advantages|Disadvantages|
|------------|-----------------------------------------------|
|Simple to implement: KNN is a straightforward algorithm to implement, especially for small to medium-sized datasets.|Computational complexity: KNN can be computationally expensive for large datasets or high-dimensional spaces.|
|Non-parametric: KNN does not require any assumptions about the underlying distribution of the data.|Sensitive to noise: KNN can be affected by noisy data, as a single outlier can significantly impact the prediction.|
|Flexible: KNN can be used for both classification and regression tasks.|Curse of dimensionality: As the number of features increases, the distance between data points becomes less meaningful, making it challenging to find reliable nearest neighbors.|

```python
import numpy as np

def knn(dataset, sample, k):
    # Calculate the distance between the sample and each data point in the dataset
    distances = []
    for data_point in dataset:
        distance = calculate_euclidean_distance(sample, data_point)
        distances.append((data_point, distance))

    # Sort the distances and select the k nearest neighbors
    distances.sort(key=lambda x: x[1])
    k_nearest_neighbors = [x[0] for x in distances[:k]]

    # Determine the predicted class or value based on the k nearest neighbors
    predicted_class = most_common_class(k_nearest_neighbors)
    return predicted_class

def calculate_euclidean_distance(sample, data_point):
    # Calculate the Euclidean distance between the sample and data point
    distance = 0
    for i in range(len(sample)):
        distance += (sample[i] - data_point[i]) ** 2
    return np.sqrt(distance)

def most_common_class(neighbors):
    # Determine the most common class among the k nearest neighbors
    classes = [neighbor[-1] for neighbor in neighbors]
    return max(set(classes), key=classes.count)
```

### Confusion matrix

A confusion matrix is a table used to evaluate the performance of a classification model by comparing its predicted values against the actual values for a dataset. It is essential as it provides insights into the model’s accuracy, precision, and recall.

example:



**Components of a Confusion Matrix**

A confusion matrix typically consists of two rows and two columns:

- ***Actual Class***: The true or actual class labels of the data instances.
* ***Predicted Class***: The predicted class labels generated by the classification model.

The four quadrants of the matrix represent the following:

- ***True Positives (TP)***: Correctly classified instances with positive actual class labels and positive predicted class labels.
* ***True Negatives (TN)***: Correctly classified instances with negative actual class labels and negative predicted class labels.
- ***False Positives (FP)***: Misclassified instances with negative actual class labels but positive predicted class labels.
* ***False Negatives (FN)***: Misclassified instances with positive actual class labels but negative predicted class labels.

**Interpretation of a Confusion Matrix**

The confusion matrix provides valuable insights into the model’s performance, including:

- ***Accuracy***: The proportion of correctly classified instances (TP + TN) divided by the total number of instances.
* ***Precision***: The proportion of true positives (TP) among all positive predicted instances (TP + FP).
- ***Recall***: The proportion of true positives (TP) among all actual positive instances (TP + FN).
* ***F1-score***: The harmonic mean of precision and recall, providing a balanced measure of both.
